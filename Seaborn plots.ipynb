```python
#################################
import seaborn as sb
%matplotlib inline
tips = sb.load_dataset('tips')    # 2 Data
flights = sb.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()       
```

```python
corr = tips.corr()           #Correlation
```

```python
sb.heatmap(corr,annot = True,cmap = 'coolwarm')       # annot for actual number
```

```python
fp = flights.pivot_table(index = 'month',columns = 'year',values = 'passengers')    ### REMAKE TABLE into MATRIX
fp.head()
```

```python
sb.heatmap(fp,cmap= 'magma',linecolor='black' ,linewidth = 0.2)   #line SEPERATION
```

```python
sb.clustermap(fp)
```

```python
sb.lmplot(x = 'total_bill',y = 'tip',data = tips,col = 'day',aspect = 0.6) # hue = 'coolwamr',markers = ['o','v'],
                 #scatter_kws={'s':60})           #### {'s': (for size)}
```

```python
g = sb.PairGrid(tips)
g.map_upper(sb.kdeplot)
```
